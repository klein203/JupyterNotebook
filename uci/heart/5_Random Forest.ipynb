## 导入工具包，预处理后的数据集

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

df = pd.read_csv("./data/process_heart.csv")
```

```python
df.shape
```

## 划分特征列和标签列

```python
X = df.drop("target", axis=1)
X.shape
```

```python
y = df.target
y.shape
```

## 划分训练集和测试集

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## 构建随机森林模型，在训练集上训练模型

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(max_depth=5, n_estimators=10, random_state=5)
model.fit(X_train, y_train)
```

## 可视化随机森林的一棵决策树

```python
len(model.estimators_)
```

```python
estimator = model.estimators_[0]
estimator
```

```python
feature_names = X_train.columns
y_train_str = y_train.astype("str")
y_train_str[y_train_str == "0"] = "no disease"
y_train_str[y_train_str == "1"] = "disease"
y_train_str = y_train_str.values
```

```python
from sklearn.tree import export_graphviz

export_graphviz(estimator, out_file="./temp/tree.dot", feature_names=feature_names, class_names=y_train_str, rounded=True, proportion=True, label="root", precision=2, filled=True)
```
