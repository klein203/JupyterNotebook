# key words
- inline shortcut: datasets.make_*?
- datasets: make_regression, make_classification, make_blobs

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

# datasets.make_*?

# (100 x 100, 100)
reg_data = datasets.make_regression()

# (1000 x 10, 100 x 2)
complex_reg_data = datasets.make_regression(n_samples=1000, n_features=10, n_informative=5, n_targets=2, bias=1.0)

# (100 x 20, 100), weights stands for the distribution of bi-classes
classification_set = datasets.make_classification(weights=[0.1])
print(np.bincount(classification_set[1]))

# clustering data
blobs_data, blobs_target = datasets.make_blobs()
plt.scatter(blobs_data[:, 0], blobs_data[:, 1], c=blobs_target)
```
