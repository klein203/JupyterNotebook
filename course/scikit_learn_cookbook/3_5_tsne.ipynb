# key words
- decomposition: PCA, TSNE, MDS, ISO

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

from sklearn.decomposition import PCA
from sklearn.manifold import TSNE, MDS, Isomap

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_pca = PCA(n_components=2).fit_transform(X)
X_tsne = TSNE(learning_rate=200).fit_transform(X)
X_MDS = MDS(n_components=2).fit_transform(X)
X_ISO = Isomap(n_components=2).fit_transform(X)

X_pca_then_tsne = TSNE(learning_rate=200).fit_transform(X_pca)

plt.figure(figsize=(20, 15))

plt.subplot(3, 2, 1, title="PCA")
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y)

plt.subplot(3, 2, 2, title="TSNE")
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y)

plt.subplot(3, 2, 3, title="MDS")
plt.scatter(X_MDS[:, 0], X_MDS[:, 1], c=y)

plt.subplot(3, 2, 4, title="ISO")
plt.scatter(X_ISO[:, 0], X_ISO[:, 1], c=y)

plt.subplot(3, 2, 5, title="PCA followed by TSNE")
plt.scatter(X_pca_then_tsne[:, 0], X_pca_then_tsne[:, 1], c=y)

plt.show()
```
