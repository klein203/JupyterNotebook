```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans, MiniBatchKMeans
from sklearn.metrics import silhouette_samples, silhouette_score, normalized_mutual_info_score, mutual_info_score, pairwise


X_blobs3, y_blobs3 = make_blobs()

distances = pairwise.pairwise_distances(X_blobs3)

# 5 of closest point to X_blobs[0][0]
distance_indexes = np.argsort(distances[0])
closest_points = X_blobs3[distance_indexes[:5]]
extreme_points = X_blobs3[distance_indexes[-5:]]

# plot
plt.figure(figsize=(10, 7))
plt.scatter(X_blobs3[:, 0], X_blobs3[:, 1], marker='.', label='All Points')
plt.scatter(X_blobs3[0, 0], X_blobs3[0, 1], color='green', marker='*', label='Chosen Point')
# sp_points[0, 0] refers to the chosen point, skip it
plt.scatter(closest_points[1:, 0], closest_points[1:, 1], color='red', marker='+', label='Closest Points')
plt.scatter(extreme_points[:, 0], extreme_points[:, 1], edgecolors='g', facecolors='none', s=80, label='Extreme Points')

plt.legend(loc='best')
```
