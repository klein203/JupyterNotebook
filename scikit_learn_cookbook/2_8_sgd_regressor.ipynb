# key words
- regressor: SGDRegressor

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

from sklearn.linear_model import SGDRegressor

X, y = datasets.make_regression(int(1e6)) #1,000,000 rows

train_set = np.random.choice([True, False], len(y), p=[.75, .25])
X_train, X_test = X[train_set], X[~train_set]
y_train, y_test = y[train_set], y[~train_set]

sgd = SGDRegressor()
# sgd = SGDRegressor(alpha=0.0001, average=False, epsilon=0.1, eta0=0.01, fit_intercept=True, l1_ratio=0.15, learning_rate='invscaling', loss='squared_loss', n_iter=5, penalty='l2', power_t=0.25, random_state=None, shuffle=True, verbose=0, warm_start=False)
sgd.fit(X_train, y_train)
y_pred = sgd.predict(X_test)

plt.figure(figsize=(12, 6))

plt.hist(y_pred - y_test, bins=50)

plt.show()
```
