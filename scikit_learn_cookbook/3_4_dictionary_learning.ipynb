# key words
- decomposition: DictionaryLearning
- matplot: 3D plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

from sklearn import decomposition

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test = X[::2], X[1::2]
y_train, y_test = y[::2], y[1::2]

dl_n3 = decomposition.DictionaryLearning(n_components=3)
X_train_transform = dl_n3.fit_transform(X_train)
# print(X_train_transform[:5])

X_test_transform = dl_n3.transform(X_test)
# print(X_test_transform[:5])

plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1, projection='3d', title="Training set")
plt.scatter(X_train_transform[:, 0], X_train_transform[:, 1], X_train_transform[:,2], c=y_train, marker='^')


plt.subplot(1, 2, 2, projection='3d', title="Testing set")
plt.scatter(X_test_transform[:, 0], X_test_transform[:, 1], X_test_transform[:,2], c=y_test, marker='^')

plt.show()
```
