# key words
- decomposition: TruncatedSVD

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from sklearn import datasets

from sklearn import decomposition

iris = datasets.load_iris()
X = iris.data
y = iris.target

svd = decomposition.TruncatedSVD(n_components=2)
X_truncsvd_n2 = svd.fit_transform(X)
print(svd.explained_variance_ratio_.sum(), svd.explained_variance_ratio_)

plt.figure(figsize=(10, 14))

plt.subplot(2, 1, 1, title="before truncated svd")
plt.scatter(X[:, 0], X[:, 1], c=y, s=40)

plt.subplot(2, 1, 2, title="after truncated svd (n=2)")
plt.scatter(X_truncsvd_n2[:, 0], X_truncsvd_n2[:, 1], c=y, s=40)

plt.show()
```
