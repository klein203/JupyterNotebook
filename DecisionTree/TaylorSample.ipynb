```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(- np.pi * 1.2, np.pi * 1.2, 100)
y = np.sin(X)

def t0(X, X0):
    return np.sin(X0) / np.math.factorial(0)

def t1(X, X0):
    return t0(X, X0) + np.cos(X0) * (X ** 1) / np.math.factorial(1)

def t2(X, X0):
    # equals to t1
    return t1(X, X0) - np.sin(X0) * (X ** 2) / np.math.factorial(2)

def t3(X, X0):
    return t2(X, X0) - np.cos(X0) * (X ** 3) / np.math.factorial(3)

def t4(X, X0):
    # equals to t3
    return t3(X, X0) + np.sin(X0) * (X ** 4) / np.math.factorial(4)

def t5(X, X0):
    return t4(X, X0) + np.cos(X0) * (X ** 5) / np.math.factorial(5)

def t6(X, X0):
    # equals to t4
    return t5(X, X0) - np.sin(X0) * (X ** 6) / np.math.factorial(6)

def t7(X, X0):
    return t6(X, X0) - np.cos(X0) * (X ** 7) / np.math.factorial(7)

plt.figure(figsize=(6, 6))
plt.plot(X, y, 'k-', label='y=sin(x)', linewidth=3)
plt.plot(X, t7(X, 0), 'y-', label='Taylor7', linewidth=2)
plt.plot(X, t5(X, 0), 'g:', label='Taylor5')
plt.plot(X, t3(X, 0), 'r:', label='Taylor3')
plt.plot(X, t1(X, 0), 'b:', label='Taylor1')
plt.legend(loc='lower right', shadow=True, fontsize='x-large')

plt.show()
```
