```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# fn = x^2 + y^2
def fn(X):
    return np.sqrt(np.power(X[:, 0], 2) + np.power(X[:, 1], 2))

x_min, x_max = -1. - .05, 1. + .05
y_min, y_max = -1. - .05, 1. + .05
plot_step = .02
X = [np.arange(x_min, x_max, plot_step), np.arange(y_min, y_max, plot_step)]
xx, yy = np.meshgrid(X[0], X[1])
z = fn(np.c_[xx.ravel(), yy.ravel()])
z = z.reshape(xx.shape)

plt.figure(figsize=(8, 8))
plt.contourf(xx, yy, z, cmap=plt.cm.RdYlBu)
plt.show()
```
