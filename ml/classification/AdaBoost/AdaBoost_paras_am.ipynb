```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
threhold = .5
x = np.linspace(.1, .9, 50)
plt.figure(figsize=(6 * 3, 6))

y = np.log((1 - x) / x)
plt.subplot(1, 3, 1)
plt.title('a_m = f(e_m) = ln((1-e_m)/e_m)', fontsize=12)
plt.plot(x[x <= threhold], y[x <= threhold], 'r-')
plt.plot(x[x > threhold], y[x > threhold], 'b:')

dy = 1 / (x - 1) - 1 / x
plt.subplot(1, 3, 2)
plt.title('da_m = f\'(e_m) = 1/(e_m-1) - 1/e_m', fontsize=12)
plt.plot(x[x <= threhold], dy[x <= threhold], 'r-')
plt.plot(x[x > threhold], dy[x > threhold], 'b:')

d2y = - 1 / (x - 1) ** 2 + 1 / x ** 2
plt.subplot(1, 3, 3)
plt.title('d2a_m = -1/((e_m - 1)**2) + 1/(e_m**2)', fontsize=12)
plt.plot(x[x <= threhold], d2y[x <= threhold], 'r-')
plt.plot(x[x > threhold], d2y[x > threhold], 'b:')
plt.show()
```
