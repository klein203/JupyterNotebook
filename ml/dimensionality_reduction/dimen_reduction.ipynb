```python
from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.preprocessing import minmax_scale
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

digits = datasets.load_digits()
# print(digits.DESCR)
```

```python
X = digits.data
y = digits.target
```

```python
img = X[0].reshape((8, 8))
img = img * 256 / 16
im = Image.fromarray(img)
im = im.resize((128, 128))
im.show()
```

```python
tsne = TSNE(n_components=2, init='pca', learning_rate='auto', random_state=0)
X_tsne = tsne.fit_transform(X)
X_tsne = minmax_scale(X_tsne)

plt.figure(figsize=(8, 8))
scatter = plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=y, marker='.')
plt.legend(*scatter.legend_elements())
plt.title('TSNE 2D plot of the digits dataset')
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
X_pca = minmax_scale(X_pca)

plt.figure(figsize=(8, 8))
scatter = plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, marker='.')
plt.legend(*scatter.legend_elements())
plt.title('PCA 2D plot of the digits dataset')
plt.show()
```

```python
tsne = TSNE(n_components=3, init='pca', learning_rate='auto', random_state=0)
X_tsne = tsne.fit_transform(X)
X_tsne = minmax_scale(X_tsne)

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=y, marker='.')
ax.legend(*scatter.legend_elements())
plt.title('TSNE 3D plot of the digits dataset')
plt.show()
```

```python
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X)
X_pca = minmax_scale(X_pca)

fig = plt.figure(figsize=(12, 12))
ax = fig.add_subplot(projection='3d')
scatter = ax.scatter(X_pca[:, 0], X_pca[:, 1], X_pca[:, 2], c=y, marker='.')
ax.legend(*scatter.legend_elements())
plt.title('PCA 3D plot of the digits dataset')
plt.show()
```
